```python
import pandas as pd
import math
import os
import zipfile

try:
    from google.colab import files

    IN_COLAB = True

except ImportError:

    IN_COLAB = False


if IN_COLAB:

    print("Upload your CSV or Excel file")

    uploaded = files.upload()

    input_file = list(uploaded.keys())[0]

    print(f"\nUploaded File: {input_file}")

else:

    input_file = input("Enter CSV or Excel file path: ").strip()

    if not os.path.exists(input_file):
        raise FileNotFoundError("File not found!")

    print(f"\nSelected File: {input_file}")


file_extension = input_file.split(".")[-1].lower()

if file_extension == "csv":

    df = pd.read_csv(input_file)

elif file_extension in ["xlsx", "xls"]:

    df = pd.read_excel(input_file)

else:

    raise ValueError("Unsupported file format. Please upload CSV or Excel file.")


total_rows = len(df)

print(f"\nTotal Rows Found: {total_rows}")


chunk_size = int(input("\nEnter number of rows per chunk: "))

if chunk_size <= 0:

    raise ValueError("Chunk size must be greater than 0")


total_chunks = math.ceil(total_rows / chunk_size)

print(f"\nTotal Chunks to be Created: {total_chunks}")


base_name = os.path.splitext(os.path.basename(input_file))[0]


output_folder = f"{base_name}_chunks"

os.makedirs(output_folder, exist_ok=True)


created_files = []


for chunk_number in range(total_chunks):

    start_idx = chunk_number * chunk_size
    end_idx = start_idx + chunk_size

    chunk_df = df.iloc[start_idx:end_idx]

    output_file = os.path.join(
        output_folder,
        f"{base_name}_chunk_{chunk_number + 1}.xlsx"
    )

    chunk_df.to_excel(output_file, index=False)

    created_files.append(output_file)

    print(f"Created: {output_file}")


zip_file_name = f"{base_name}_chunks.zip"


with zipfile.ZipFile(zip_file_name, 'w') as zipf:

    for file in created_files:

        zipf.write(file)


print("\nZIP file created successfully!")


if IN_COLAB:

    print(f"\nDownloading: {zip_file_name}")

    files.download(zip_file_name)

else:

    print(f"\nZIP file saved locally: {zip_file_name}")
```
